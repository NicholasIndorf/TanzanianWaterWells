```python
import pandas as pd
import numpy as np
import geopandas as gpd
from scipy.spatial import cKDTree
from shapely.geometry import Point
import distance_function
```

```python
gdf = gpd.read_file('GISdata/X_y_train_gdf.shp')
gdf.head()
```

```python
gdf_sample = gdf.head(10)
gdf_sample.info()
```

```python
gdf_DeS = gdf[gdf['region']=='Dar es Salaam']
gdf_DeS.reset_index(inplace = True)
```

```python

results = []
for x in range(len(gdf_DeS)):
    gdf_a = gdf_DeS.iloc[x]
    nA = [[gdf_a.geometry.x, gdf_a.geometry.y]]
    print(nA)
    
    gdf_b = gdf_DeS.drop(x)
    nB = np.array(list(gdf_b.geometry.apply(lambda x: (x.x, x.y))))


    btree = cKDTree(nB)

    dist, idx = btree.query(nA, k=3) #k value gives you the number of nearest points to return


    dist = list(dist[0])
    idx = list(idx[0])
    status = []
    ids = []
    orig_row_id = gdf_DeS.iloc[x]['id']
    
    for x in idx:
        status.append(gdf_b.iloc[x]['status_gro'])
        ids.append(gdf_b.iloc[x]['id'])

    results_dict = {'original row id':orig_row_id,'ids':ids, 'statuses':status, 'dist':dist}
    results.append(results_dict)

```

```python


results_df = pd.DataFrame(results)
results_df.head(10)
```

```python
gdf_w_results = pd.merge(gdf_DeS, results_df, left_on='id',right_on='original row id')

gdf_w_results.info()
```
